### Câu 1: Thử thách phân tích thừa số #1

```python
# %pip install gmpy2
import gmpy2
from gmpy2 import mpz, mpfr

# Giá trị của N được cho
N = mpz(
    "179769313486231590772930519078902473361797697894230657273430081157732675805505620686985379449212982959585501387537164015710139858647833778606925583497541085196591615128057575940752635007475935288710823649949940771895617054361149474865046711015101563940680527540071584560878577663743040086340742855278549092581"
)

# Tính căn bậc hai của N
sqrt_N = gmpy2.isqrt(N)

# Tìm A,x
A = sqrt_N + 1
x = gmpy2.isqrt(A**2 - N)

# Tính thừa số p, q
p = A - x
q = A + x

print("N =", N)
print("A =", A)
print("P =", p)
print("q =", q)
print("P là số nguyên tố: ", gmpy2.is_prime(p))
print("q là số nguyên tố: ", gmpy2.is_prime(q))
print("p * q =", p * q)
print("Kiểm tra p * q = N ", p * q == N)
```

### Câu 3: Thử thách phân tích thừa số #3

##### Theo tính toán cá nhân, (3p+2q) sẽ rất gần với căn 24N. Dat A = (3p+2q) x = (3p-2q). Vi p,q là số nguyên tố nên A, x là số nguyên


#### Baì toán sẽ đưa về căn 24N với A = (3p+2q) để tránh xử lý (3p + 2q)/2 có giá trị thực .05f

```python
N = mpz(
    "720062263747350425279564435525583738338084451473999841826653057981916355690188337790423408664187663938485175264994017897083524079135686877441155132015188279331812309091996246361896836573643119174094961348524639707885238799396839230364676670221627018353299443241192173812729276147530748597302192751375739387929"
)

sqrt_24N = gmpy2.isqrt(24 * N)
A = sqrt_24N + 1

X = gmpy2.isqrt(A**2 - 24 * N)

print("Dự đoán x: ")
print("(A + X) % 3 =", (A + X) % 3)
print("(A + X) % 2 =", (A + X) % 2)
print(
    "Do A + X chia 3 dư 2, chia hết cho 2 nên ở đây x = 2q -3p để A + x = 4q va A - x = 6p"
)
# Do A + X chia 3 du 2, chia het cho 2 nen o day x = 2q -3p de A + x = 4q va A - x = 6p

# Tính p, q
q = (A - X) // 6
p = (A + X) // 4
print("Kết quả: ")
print("A =", A)
print("p =", p)
print("q =", q)
print("P là số nguyên tố: ", gmpy2.is_prime(p))
print("q là số nguyên tố: ", gmpy2.is_prime(q))

print("Kiểm tra lại, p * q = N :", p * q == N)
```
